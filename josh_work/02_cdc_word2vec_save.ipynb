```python
docs = list(df['text_edit'])
```

```python
%%time
mod = gensim.models.Word2Vec(
        docs,
        size=100,
        window=5,
        min_count=3,
        iter=20,
        sg=1)
```

Investigate similarity on common words

```python
w1 = 'arm'
mod.wv.most_similar(w1)
```

```python
for i in range(50):
    w1 = mod.wv.index2word[i]
    print(f"\n-----{w1}---------------------------------")
    topw = mod.wv.most_similar(w1,topn=3)
    topw = [w[0] for w in topw]
    print(topw)
```

Least Common

```python
vocab_size = len(mod.wv.vocab)
mod.wv.index2word[vocab_size - 10:]
checkn = 50
for i in range(checkn):
    w1 = mod.wv.index2word[vocab_size - i - 1]
    print(f"\n-----{w1}---------------------------------")
    topw = mod.wv.most_similar(w1,topn=3)
    topw = [w[0] for w in topw]
    print(topw)
```
